```python
import lxml
import scraperwiki
import re
import itertools
import csv
```

```python
f=open("../data/raw/listofdeaths.pdf")
pdf=f.read()
f.close()
```

```python
x=scraperwiki.pdftoxml(pdf)
r=lxml.etree.fromstring(x)
```

```python
import pprint
```

```python
pprint.pprint(x)
```

```python
pages=r.xpath("//page[@number<35]")
```

```python
[i.text for i in pages[0].xpath("./text[@font=0]")]
```

```python
def isdate(s):
    return re.match("[0-9]{2}/[0-9]{2}/[0-9]{2}",s) or re.match("in .*?[0-9]{2}",s)

def split_on_date(i):
    rr=[]
    r=[i[0]]
    for j in i[1:]:
        if isdate(j):
            rr.append(r)
            r=[]
        r.append(j)
    rr.append(r)
    return rr
    
        
def split_entries(x,y):
    """ split joined entries (happens if the first column is two digits) """
    if re.match("^[0-9]{2} .*?",y):
        x.extend(y.split(" ",1))
    else:
        x.append(y)
    return x
        
def process_page(p):
    e=[i.text for i in p.xpath("./text[@font=0]")]
    e=e[2:]
    #e=[i for i in split_on_date(reduce(split_entries,e,[]))]
    return split_on_date(reduce(split_entries,e,[]))
    
```

```python
data=reduce(lambda x,y: x+y,[process_page(p) for p in pages])
data
```

```python
f=open("../data/listofdeath.csv","wb")
w=csv.writer(f)
w.writerow(["date","count","name","origin","cause of death","source"])
for d in data:
    w.writerow([i.encode("utf-8") for i in d])
f.close()
```
